```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 200):
    review = re.sub('[^a-zA-Z]', ' ', df['Reviews'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
review_vectorizer = vectorizer.fit_transform(corpus)
```

```python
print(vectorizer.get_feature_names())
```

```python
review_vectorizer.toarray().shape
```

```python
print(review_vectorizer.toarray())
```

```python
df.drop('Reviews', axis = 1, inplace = True)
```

```python
rev = pd.DataFrame(review_vectorizer.toarray())
```

```python
rev
```

```python
df = pd.concat([df, rev], axis = 1)
```

```python
df.head(40)
```

```python
X = df.drop('Label', axis = 1)
```

```python
y = df['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 200, criterion= 'entropy', random_state = 100)
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_predict = classifier.predict(X_test)
```

```python

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```
